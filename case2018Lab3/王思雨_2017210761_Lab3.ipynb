## Part2：China Daily 足球新闻爬虫

### 2.1爬取每条足球新闻url以及标题函数

```python
def Crawl(urls):
    results={}
    driver = webdriver.PhantomJS()            #无浏览器自动化过程，节省内存使用，简化操作，提高速度
    for url in urls:
        driver.get(url)
        page = driver.page_source
        soup = BeautifulSoup(page, 'html.parser')
        span1 = soup.find_all('div', class_='mb10 tw3_01_2 ')         
        for l in span1:
            title =l.find('h4').get_text()    #找到歌曲title并储存
            url = l.find('a')['href']         #拼接歌曲的url
            print (title,url)                 #输出title以及歌曲链接
            results.setdefault(title,url)
    return results 
```

### 2.2根据url爬取每条足球新闻文本信息函数

```python
def Crawl_lyrics(url):
    driver = webdriver.PhantomJS()
    driver.get(url)
    time.sleep(2)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    span2 = soup.find_all('p') 
    span2 = [item.get_text() for item in span2[:-2] ]
    str = "\n"
    text = str.join( span2 )
    print ('success')
    driver.quit()
    return text                              #抛出文本内容
```

### 2.3写出新闻文本和标题到excel表格函数

```python
def write_to_excel(results,path):
    workbook = xlwt.Workbook(encoding='utf-8')       #打开一个工作簿，编码为utf-8
    worksheet = workbook.add_sheet(u'china_daily')
    worksheet.write(0, 0, 'title')
    worksheet.write(0, 1, 'text')
    flag = 1
    for key, value in results.items():
        worksheet.write(flag, 0, key)
        url = value
        text = Crawl_lyrics(url).split(':')[-1]
        worksheet.write(flag, 1, text)
        flag += 1
    workbook.save(path)
```

```python
site=[]
for i in list(range(34)):
    web = "http://www.chinadaily.com.cn/sports/soccer/page_"+str(i+1)+".html"
    site.append(web)        
football = Crawl(site)
write_to_excel(football,'D:/bigdatahw/Case contest/data/football.xls')
```

```python
class loadFolders(object):          # 文件夹打开函数
    def __init__(self, par_path):
        self.par_path = par_path
    def __iter__(self):
        for file in os.listdir(self.par_path):
            file_abspath = os.path.join(self.par_path, file)
            if os.path.isdir(file_abspath):  
                yield file_abspath
```

```python
class loadFiles(object):            # txt文档打开函数
    def __init__(self, par_path):
        self.par_path = par_path
    def __iter__(self):
        folders = loadFolders(self.par_path)
        for folder in folders:              
            catg = folder.split(os.sep)[-1]
            for file in os.listdir(folder):     
                file_path = os.path.join(folder, file)
                if os.path.isfile(file_path):
                    this_file = open(file_path, 'rb')
                    content = this_file.read().decode('utf8')
                    yield catg, content
                    this_file.close()       
```

### 4.2足球新闻文本数据处理函数
+ 还原英文省略词汇函数
  + replacement_patterns定义tih替换模式
  + RegexpReplacer类用于实例化替换对象

```python
replacement_patterns = [
(r'won\'t', 'will not'),
(r'can\'t', 'cannot'),
(r'i\'m', 'i am'),
(r'ain\'t', 'is not'),
(r'(\w+)\'ll', '\g<1> will'),
(r'(\w+)n\'t', '\g<1> not'),
(r'(\w+)\'ve', '\g<1> have'),
(r'(\w+)\'s', '\g<1> is'),
(r'(\w+)\'re', '\g<1> are'),
(r'(\w+)\'d', '\g<1> would')
]

class RegexpReplacer(object):
    def __init__(self, patterns=replacement_patterns):
        self.patterns = [(re.compile(regex), repl) for (regex, repl) in patterns]
    def replace(self, text):
        s = text
        for (pattern, repl) in self.patterns:
            (s, count) = re.subn(pattern, repl, s)
        return s
```

+ 删除重复字符函数

```python
class RepeatReplacer(object):
    def __init__(self):
        self.repeat_regexp = re.compile(r'(\w*)(\w)\2(\w*)')
        self.repl = r'\1\2\3'
    def replace(self, word):
        if wordnet.synsets(word):
            return word
        repl_word = self.repeat_regexp.sub(self.repl, word)
        if repl_word != word:
            return self.replace(repl_word)
        else:
            return repl_word
```

+ 去除空白符函数

```python
def rm_char(text):
    text = re.sub('\u3000', '', text)    #全角的空白符
    return text
```

+ 去除数字、非英文字符、停用词、转换成小写、词形还原等文本整体处理函数

```python
def rm_tokens(words):  # 去掉一些停用词和完全包含数字的字符串
    words = words.encode("utf-8").decode("utf-8")
    words = re.sub("\d","",words)
    Reg = RegexpReplacer()                           #删除缩写
    words_seg = [re.sub(u'\W', "", Reg.replace(i)) for i in nltk.word_tokenize(words)] 
    space_len = words_seg.count(u"")
    for i in list(range(space_len)):
        words_seg.remove(u'')
    filtered = [w.lower() for w in words_seg if w not in stopwords.words('english')] 
    wnl = WordNetLemmatizer()
    stemmer = PorterStemmer()
    #Rep = RepeatReplacer()
    lemmatized = [wnl.lemmatize(w) for w in filtered] #词形还原
    stems = [stemmer.stem(w) for w in lemmatized]     #词干提取
    #final = [Rep.replace(w) for w in stems]          #删除重复字符
    return " ".join(stems)
```

+ 文本转单词主函数

```python
def convert_doc_to_wordlist(str_doc):
    sent_list = str_doc.split('\n')
    sent_list = map(rm_char, sent_list)  # 去掉一些字符，例如\u3000
    word_list = [rm_tokens(part) for part in sent_list]  # 分词
    word2list = " ".join(word_list)
    return word2list
```

### 4.3新闻文本处理主函数

```python
files = loadFiles(path_doc_root)
football = pd.read_excel('D:/bigdatahw/Case contest/data/football.xls')
words=[]
for i, msg in enumerate(files):
    file = msg[1]
    word_list = convert_doc_to_wordlist(file)    
    words.append(word_list)
news=pd.DataFrame({'title':football["title"].tolist(),'words':words})
news.head(5)
```

### 4.4新闻文本分词函数

```python
def get_word_tf_mat(sparce_mat):
    arr = sparce_mat.toarray().T
    word_tf = pd.DataFrame(arr,index = vectorizer.get_feature_names())
    word_tf["tf"] = np.sum(arr,axis=1)
    word_tf = word_tf.sort_values(by="tf",ascending=False)
    return word_tf.T
```

### 4.5新闻文本分词主函数

```python
content = [t for t in news["words"]]
vectorizer=CountVectorizer(stop_words='english',min_df=10)  #利用机器学习库进行分词，只选择词频大于5的词
count_word = vectorizer.fit_transform(content)
word_tf = get_word_tf_mat(count_word)
word_tf.to_csv("word_tf.csv")     
data = pd.read_csv("word_tf.csv")                           
l = football["title"].tolist()
data["Unnamed: 0"] = l.append("tf")
data.rename(columns={'Unnamed: 0':'title'}, inplace = True)  #重命名列名
data.ix[:,:10].head(10)
```
